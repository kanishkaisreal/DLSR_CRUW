```python
import os
import numpy as np
import math
```

```python
yolo_file_path = "/home/vr-lab/Asish/8th Sem Radar Project/Datasets/CRUW_SR_Annotation_Yolo_format"
```

```python
car = 0
ped = 0
cyc = 0
cari = 0 
pedi = 0
cyci = 0

fname = '/home/vr-lab/Asish/8th Sem Radar Project/Object Detection/ssd-tf2-master/dataset/VOC2007_SR/ImageSets/train.txt'
train_dne = []

with open (fname,'r') as fl:
    file_list = fl.readlines()
    print("Total number of images : " ,len(file_list))
    for file in file_list:
        file_yolo = yolo_file_path + "/" + file[-36:-5] + "txt"
#         print(file[-36:-5])
        if not os.path.exists(file_yolo):
#             print(file_yolo)
            train_dne.append(file)
        else:
            with open(file_yolo, 'r') as f:
                lines = f.readlines()
                classes = []
                for line in lines:
                    classes.append(int(line.split(" ")[0]))
                if 1 in classes:
                    cari = cari + 1
                    car = car + classes.count(1)
                if 2 in classes:
                    cyci = cyci + 1
                    cyc = cyc + classes.count(2)
                if 3 in classes:
                    pedi = pedi + 1
                    ped = ped + classes.count(3)
            
print("Total no of cars : ", car, " in ", cari, " images")
print("Total no of cyclists : ", cyc, " in ", cyci, " images")
print("Total no of pedestrians : ", ped, " in ", pedi, " images")
```
